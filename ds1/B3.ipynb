```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
A=pd.read_csv("airquality1.csv",delimiter=',')
```

# A)Data Cleaning
Checking Missing Values in Dataframe

```python
A.isnull().sum()
```

# Replacing the null values


```python
A["Ozone"]=A["Ozone"].fillna(A["Ozone"].mean())
A["Solar.R"]=A["Solar.R"].fillna(A["Solar.R"].mean())
```

```python
A["humidity"].mode()
```

```python
A["humidity"]=A["humidity"].fillna('high')
```

```python
A
```

```python
A.isnull().sum()
```

```python
A.drop('Unnamed: 0',axis=1,inplace=True)
```

# 2.Bar plot


```python

def addlabels():
 for i in range(len(A.iloc[0:21,3])):
    plt.text(i,A.iloc[0:21,3][i],A.iloc[0:21,3][i],ha='center')
 
y=np.arange(len(A.iloc[0:21,3]))
plt.figure(figsize=(10,6))
addlabels()
plt.bar(x=y,height=A.iloc[0:21,3],tick_label=y,color="dodgerblue")
plt.title('Wind measure for 1st-20th May')
plt.xlabel('Days')
plt.ylabel('Wind')
plt.show()
```

# 3.Box_plot

```python
A.boxplot(column=['Ozone','Wind','Temp','Solar.R'],grid=True)
plt.title('Box-Plot')
plt.suptitle('Temperature')
plt.ylabel('range')
plt.savefig('plot3.png')
```

# 4.Histogram

```python
# plt.hist(x=A['Temp'],bins=10,color="blue")
plt.figure(figsize=(10,6))
plt.hist(x=A['Temp'],bins=20,color="Red",label="Temp")
plt.title("Temperature count")
plt.xlabel("Temperature")
plt.ylabel("count")
plt.plot()
```
